```python
import sys
from pathlib import Path
import pandas as pd
import numpy as np
from datetime import datetime
from sklearn.feature_extraction.text import TfidfVectorizer, ENGLISH_STOP_WORDS

PROJECT_ROOT = Path("..").resolve()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

FINAL_PATH = PROJECT_ROOT / "results/reports/onet_occupations.csv"
assert FINAL_PATH.exists(), f"File not found: {FINAL_PATH}"

df = pd.read_csv(FINAL_PATH)
```

```python
flat = (
    df.groupby("Title")["Description"]
      .apply(lambda x: ", ".join(sorted(set(
          ", ".join(x.dropna()).split(", ")
      ))))
      .reset_index()
)

df.head()
```

```python
OUT_PATH = PROJECT_ROOT / "results/reports/onet_title_flat_keywords.csv"

flat.to_csv(OUT_PATH, index=False)

print("Saved to:", OUT_PATH)
```
